### Заменим пропуски

```python
df['fr_id'] = df['fr_id'].fillna(0)
df['route_type'] = df['route_type'].fillna(2)
```

### Приведение типов

```python
df['fr_id'] = df['fr_id'].astype('int32')
df['route_type'] = df['route_type'].astype('int32')
```

### Восполним пропущенное расстояние

```python
le = LabelEncoder()
list1 = list(df['st_code_rsv'])
list1.extend(list(df['st_code_snd']))
le.fit(list1)
df['st_code_snd'] = le.transform(df['st_code_snd'])
df['st_code_rsv'] = le.transform(df['st_code_rsv'])
```

```python
list_from = list(df['st_code_rsv'])
list_to = list(df['st_code_snd'])
list_rout = []

for i in range(len(list_from)):
    list_rout.append(str(min(list_from[i], list_to[i])) + str(max(list_from[i], list_to[i])))


df['route_code'] = list_rout
```

```python
df['distance'] = df['distance'].fillna(df.groupby('route_code')['distance']
                                             .transform("median"))
df['distance'] = df['distance'].fillna(0)
```

### Закодируем `route_type`  

```python
df['rout_t1'] = df['route_type'].replace({2:0,3:0,4:0})
df['rout_t3'] = df['route_type'].replace({1:0,2:0,3:1,4:0})
df['rout_t4'] = df['route_type'].replace({1:0,2:0,3:0,4:1})
```

### Заполним пропуски в виде собственности

```python
df['vidsobst'] = df['vidsobst'].fillna(102)
```

### Заполним пропуски в хар вагона

```python
df['common_ch']  = df['common_ch'].fillna(df['common_ch'].median())
```

## Найдем скорость

```python
df['speed'] = round(df['distance']/df['y'],0)
```

## Cформируе из скорости сущности

```python
speed_by_common_ch =  dict(df.groupby('common_ch').speed.median())
df['speed_by_common_ch']  = df['common_ch'].replace(speed_by_common_ch)
df['y_err_common']  = round(df['distance']/df['speed_by_common_ch'] ,0)
```

```python
speed_by_rod =  dict(df.groupby('rod').speed.median())
df['speed_by_rod']  = df['rod'].replace(speed_by_rod)
df['y_err_rod']  = round(df['distance']/df['speed_by_rod'] ,0)
```

```python
speed_by_route_code =  dict(df.groupby('route_code').speed.median())
list1 = list(df['route_code'])
df['speed_by_common_ch']  = list1.replace(speed_by_route_code)
#df['y_err_common']  = round(df['distance']/df['speed_by_common_ch'] ,0)
```

# Машинное обучение

```python
from sklearn import linear_model
from sklearn.linear_model import Lasso, Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor as XGB

import math
from catboost import CatBoostRegressor

from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV

from sklearn.preprocessing import StandardScaler 

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
X = df.drop(['speed',
             'y'], axis=1)[:1000000]
y= df[['y']][:1000000]

#'date_depart_year',
 #            'st_code_snd', 'st_code_rsv', 'speed_by_rod',
#            'route_code', 'route_type', 'speed', 'speed_by_common_ch',
```

```python
X
```

### Разделим датасет на тренировочную и тестовую выборки

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=0)
```

### Стандартизируем данные методом StandartScaler


```python
scaler = StandardScaler()
scaler.fit(X)
X_train_st = scaler.transform(X_train)
X_test_st = scaler.transform(X_test)
```

### Функция расчета ошибки

```python
def make_prediction(m, X_train, y_train, X_test, y_test):
    model = m
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print(
        'RMSE:{:.2f}'.format(
            math.sqrt(mean_squared_error(y_test, y_pred)),
            #cross_val_score(model, X_train, y_train, cv=5).mean()
        ))
    return model
    
```

### Проверка моделей

#### Лассо регрессия

```python
print('Lasso: ')
Lasso = make_prediction(Lasso(), X_train_st, y_train, X_test_st, y_test)
```

#### Ридж регрессия

```python
print('Ridge: ')
Ridge = make_prediction(Ridge(), X_train_st, y_train, X_test_st, y_test)
```

#### Дерево решений

```python
%%time
print('Дерево решений: ')
m = DecisionTreeRegressor(max_depth=13)
DecisionTreeRegressor = make_prediction(m, X_train_st, y_train, X_test_st, y_test)
```

#### XGB

```python
from xgboost import XGBRegressor as XGB
```

```python
%%time
print('XGB: ')
m = XGB(max_depth=6, learning_rate=1, n_estimators=26)
XGB = make_prediction(m, X_train_st, y_train, X_test_st, y_test)
```

```python
print('Cat: ')
cat = make_prediction(CatBoostRegressor(), X_train_st, y_train, X_test_st, y_test)
```

```python
predict_df=pd.DataFrame({'XGB':XGB.predict(X_test_st),
                       "Cat": cat.predict(X_test_st),
                         "Tree": DecisionTreeRegressor.predict(X_test_st)
                        }) 
```

```python
predict_df['mean_predict'] = (predict_df['XGB']+predict_df['Cat']+predict_df['Tree'])/3
```

```python
math.sqrt(mean_squared_error(y_test, predict_df['mean_predict']))
```

- Без типа собственности XGB: **RMSE:80.94**
- С типом собственности и заменили пропуски на 102 **RMSE:80.56**
- Без типа собственности XGB: **RMSE:80.94**
- Без типа собственности  XGB(max_depth=6, learning_rate=1): **RMSE:79.27** 
- XGB со всеми характеристиками: **RMSE:79.43**


- Без типа собственности Сат: **RMSE:77.69**
- С типом собственности и заменили пропуски на 102 Cat: **RMSE:77.10**
- С типом собственности и заменили пропуски на 102 Cat: **RMSE:77.10**
- Сat со всеми характеристиками: **RMSE:73.79**

- XGB(max_depth=6, learning_rate=1) без `vidsobst` + Сат с `vidsobst`:**RMSE:76.07**
- XGB(max_depth=6, learning_rate=1,  n_estimators=26)  + Cat со всеми характеристиками: **RMSE:75.00**
